```python
# Population analysis
data = pd.read_csv('countries.csv')
```

```python
data
```

```python
# Compare US and Canada population
```

```python
us = data[data.country == "United States"]
```

```python
canada = data[data.country == "Canada"]
```

```python
plt.plot(us.year, us.population / 10 ** 6)
plt.plot(canada.year, canada.population / 10 ** 6)
plt.title("US vs Canada Population")
plt.xlabel("Year")
plt.ylabel("Population in millions")
plt.legend(["US", "Canada"])
plt.show()
```

```python
(us.population / us.population.iloc[0]) * 100
```

```python
# Percentage growth
plt.plot(us.year, (us.population / us.population.iloc[0]) * 100)
plt.plot(canada.year, (canada.population / canada.population.iloc[0]) * 100)
plt.title("US vs Canada Population Percentage Growth")
plt.xlabel("Year")
plt.ylabel("Population Growth (%)")
plt.legend(["US", "Canada"])
plt.show()
```
